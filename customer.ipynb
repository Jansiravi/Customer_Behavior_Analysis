### Handling Missing Values using Median

```python
df["Review Rating"]=df["Review Rating"].fillna(df["Review Rating"].median())
```

### Verification of Missing Values

```python
df.isnull().sum()
```

```python
df.columns=df.columns.str.lower()
df.columns=df.columns.str.replace(' ','_')
df.columns
```

```python
df1=df.rename(columns={'purchase_amount_(usd)':'purchase_amount'},inplace=True)
df.columns
```

```python
labels=['Young Adult','Adult','Middle-aged','Senior']
df['age_group']=pd.qcut(df['age'],q=4,labels=labels)
```

### Create a Column age_group

### Create Column purchase_frequency_days

```python
#create column purchase_frequency_days

frequency_mapping={'Fortnightly':14,
                   'Weekly':7,
                   'Monthly':30,
                   'Quarterly':90,
                   'Bi-Weekly':14,
                   'Annually':365,
                   'Every 3 Months':90
                   }
df['purchase_frequency_days']=df['frequency_of_purchases'].map(frequency_mapping)
```

## Key Insights

#### -> Male customers represent the maority of shoppers in this dataset.
#### -> Clothing is the most purchased Category and Outerwear has the lowest demand.
#### -> Customers spend slightly more on footwear and clothing compared to other categories.
#### -> Paypal is the most used payment method
#### -> Purchases are evenly distributed across seasons,meaning demand is stable year-round.

## Business Recommendations

####  1. Focus more marketing on Clothing
              -> It has the highest purchase volume.
####  2. Promote Outerwear

####  3. Offer Paypal promotions

####  4. Target repeat customers
